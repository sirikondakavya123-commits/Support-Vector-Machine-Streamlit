```python
#support vector classifier implementation
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
print(x)
```

```python
y
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
#linear  kernel implementation
svc=SVC(kernel="linear")
svc.fit(x_train,y_train)
svc.coef_
y_pred=svc.predict(x_test)
from sklearn.metrics import classification_report,confusion_matrix
print("classification_report:\n",classification_report(y_test,y_pred))
print("confusion_matrix:\n",confusion_matrix(y_test,y_pred))
```

```python
#rbf kernel implementation
rbf=SVC(kernel="rbf")
rbf.fit(x_train,y_train)
y_pred1=rbf.predict(x_test)
print("classification_report for rbf:\n",classification_report(y_test,y_pred1))
print("confusion_matrix for rbf:\n",confusion_matrix(y_test,y_pred1))
```

```python
#polynomial kernel implementation
poly=SVC(kernel="poly")
poly.fit(x_train,y_train)
y_pred2=poly.predict(x_test)
print("classification_report for polynomial:\n",classification_report(y_test,y_pred2))
print("confusion_matrix for polynomial:\n",confusion_matrix(y_test,y_pred2))
```

```python
#sigmoid kernel implementation
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(x_train,y_train)
y_pred3=sigmoid.predict(x_test)
print("classification_report for sigmoid:\n",classification_report(y_test,y_pred3))
print("confusion_matrix for sigmoid:\n",confusion_matrix(y_test,y_pred3))
```

```python
#hyperparameter tuning using SVC
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf']}
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,verbose=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

* c is regularization parameter
* gamma- influence of a single data point
* 5(c)*5(gamma)*1(kernel)=25 model

```python
y_pred4=grid.predict(x_test)
print("classification_report for hyperparameter tuning:\n",classification_report(y_test,y_pred4))
print("confusion_matrix for hyperparameter tuning:\n",confusion_matrix(y_test,y_pred4))
```

* GridSearchCV is used to perform hyperparameter tunning
* it check all combis of given parameters using CV and performance...
